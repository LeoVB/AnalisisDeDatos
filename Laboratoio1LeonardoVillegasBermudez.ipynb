```python
print("Ejercicio #1 Calcular distancia")
print("Ingrese la velocidad y el tiempo de la unicad movil")
v=float(input("Ingrese la velocidad:"))
t=float(input("Ingrese el tiempo: "))
print("La distancia es: "+str(v*t))

```

```python
print("Ejercicio #2 Calcular promedio")
nota=float(input("Ingrese la nota 1: "))
nota2=float(input("Ingrese la nota 2: "))
nota3=int(input("Ingrese la nota 3: "))
print("El promedio de las notas es: "+str((nota+nota2+nota3)/3))
```
